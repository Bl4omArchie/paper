## Fonction exponentielle

Définition: La fonction exponentielle possède deux importantes caractéristiques:
- L'exponentielle est l'unique fonction f(x) définie et dérivable sur $ \mathbb{R} $ telle que f(0) = 1 ainsi que f'(x) = f(x) $\forall \; x$.
- L'exponentielle est l'unique fonction f(x) définie et dérivable sur $ \mathbb{R} $ telle que t f(x + y) = f(x) * f(y) pour tout $\forall \; x , \;y $.

Cette fonction exponentielle se note: exp(x) ou plus courament e^x

### Résumons:
        exp(0) = 1 | exp'(x) = exp(x) | exp(x + y) = exp(x) * exp(y)
